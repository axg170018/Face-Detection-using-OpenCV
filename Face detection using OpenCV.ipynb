#### With this short notebook I will try to explain you the basics of OpenCV  which you can install using " pip install opencv-python " .We will provide an image path and a cascade path which is Haar Cascade- a machine learning object detection algorithm used to identify objects in an image or video provided by OpenCV.

```python
import cv2
#Give the path to your image
imagePath = "/Users/akashgupta/Desktop/tableau datasets/menVSwomen/crick.png"
cascPath = "haarcascade_frontalface_default.xml"
```

```python
# Create the haar cascade
faceCascade = cv2.CascadeClassifier(cascPath)
```

#### Once we create the face Cascade now we will read the image we want to use to detect faces.
#### This is now loading the face cascade into memeory so that it can be used.
#### Do not get confused Cascade is just an XML file which has been designed and contains data to detect faces

```python
# Read the image
image = cv2.imread(imagePath)
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

#### As you can notice we are now converting the picture to Grey Scale
#### We can also use the one below just to change BGR to RGB but this does not have much difference(atleast not in my case)

```python
# Read the image 2
image = cv2.imread(imagePath)
gray = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

#### Now its time to buckle up and create magic.We will now Try to detect faces in the image.
#### detectMultiScale function is the one responsible to detect real faces and will be a major part in our code.As you might have noticed already I am calling the function on object face cascade that is why it is detecting faces.Easy...... right?
#### Now let me walk you through the function step by step:
### We can see the first option which is telling that it's a grayscale image.
### Second option is the scaleFactor-Imagine the concept of perspective view where you standing closer makes your face look big but a person standing far has a face size smaller than yours.This is what scaleFactor compensates for.
### Just to generalize the concept here, a moving window is used by the detection algorith to detect objects.MinNeighbors basically defines how many objects are detected near the existing one before it declares that the face has been found.
### minSize gives the size of each window.This parameter determines how small size you want to detect. You decide it! Usually, [30, 30] is a good start for face detection.
### flags is the Parameter with the same meaning for an old cascade as in the function cvHaarDetectObjects. It is not used for a new cascade.

## Note: You have to tweak these values to get your perfect set of parameters to detect faces.I used these parameters after understanding my picture.

```python
# Detect faces in the image
faces = faceCascade.detectMultiScale(
    gray,
    scaleFactor=1.2,
    minNeighbors=2,
    minSize=(30, 30),
    flags = cv2.CASCADE_SCALE_IMAGE
)
```

```python
# Let us see how many faces we found in our Picture!
```

```python
print("Found {0} faces!".format(len(faces)))
```

```python
# Draw a rectangle around the faces
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
```

## We can see that the above function returns 4 values. 'x' and 'y' are the respective locations of the rectangle and 'w' and 'h' are the rectangles width and height.
## We have a built-in rectangle() function whichh we will draw using these values above.

```python
print("Found {0} faces!".format(len(faces)))
```
